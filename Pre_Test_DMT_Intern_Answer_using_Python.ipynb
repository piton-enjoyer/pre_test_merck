```python
def split_dataframe(df, sep=';'):
  new_columns = []
  for col in df.columns:
    new_columns.extend(col.split(sep))

  split_data = {}
  for col in df.columns:
    split_values = df[col].astype(str).str.split(sep, expand=True)
    for i in range(split_values.shape[1]):
        col_name = f"{col.split(sep)[i]}_{i}" if i < len(col.split(sep)) else f"{col}_{i}"
        while col_name in split_data:
            col_name = f"{col_name}_dup"
        split_data[col_name] = split_values[i]

  new_df = pd.DataFrame(split_data)

  assigned_columns = []
  col_index = 0
  for col in df.columns:
      split_header_parts = col.split(sep)
      for part in split_header_parts:
          if col_index < len(new_df.columns):
              assigned_columns.append(part.strip())
              col_index += 1
          else:
              break
      if col_index >= len(new_df.columns):
          break

  while col_index < len(new_df.columns):
       assigned_columns.append(new_df.columns[col_index])
       col_index += 1

  new_df.columns = assigned_columns[:len(new_df.columns)]

  return new_df
```

```python
# Split data using split_dataframe function

split_df = split_dataframe(df1, sep=';')
split_df = split_df.rename(columns={'Date / Tanggal;Customer Name / Nama Pelanggan;Product Code / Kode Barang;Description / Nama Barang_4': 'Quantity'})
```

```python
# Change data format (datetime and integer)

split_df['Date / Tanggal'] = pd.to_datetime(split_df['Date / Tanggal'], errors='coerce')
split_df['Quantity'] = split_df['Quantity'].astype(int)
```

```python
split_df['Customer Name / Nama Pelanggan'] = split_df['Customer Name / Nama Pelanggan'].replace({'PERSEROAN TERBATAS ALU AKSARA PRATAMA': 'ALU AKSARA PRATAMA, PT',
                                                                                                 'BAPAK ALI JAPTONO': 'ALI JAPTONO'})
```

```python
split_df = split_df.sort_values(by='Date / Tanggal').reset_index(drop=True)
```

```python
# Mapping 'Harga Jual'

merged_df = split_df.merge(df2, left_on='Product Code / Kode Barang', right_on='Product Code / Kode Barang', how='left')
merged_df.head()
```

```python
# Calc 'Total Harga Jual'

merged_df['Total Harga Jual'] = merged_df['Quantity'] * merged_df['Harga Jual']
```
